```python
#making a basic boken line graph

#importing bokeh
from bokeh.plotting import figure
from bokeh.io import output_file, show

#prepare some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

#prepare the output file 
output_file("Line.html")

#creates figure object
f = figure()

#create line plot
f.line(x, y)

#write the plot in the figure object
show(f)
```

```python
#making a basic boken line graph

#importing bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prepare some data
df = pandas.read_csv("data.csv")
x = df["x"]
y = df["y"]

#prepare the output file 
output_file("Line_from_csv.html")

#creates figure object
f = figure()

#create line plot
f.line(x, y)

#write the plot in the figure object
show(f)
```

```python
#making a boken  graph with women bachelors

#importing bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prepare some data
df = pandas.read_csv("bachelors.csv")
x = df["Year"]
y = df["Engineering"]

#prepare the output file 
output_file("Line_from_bachelors.html")

#creates figure object
f = figure(plot_width=500,plot_height=400, tools='pan')
f.title.text="Cool Data"
f.title.text_color="Gray"
f.title.text_font="times"
f.title.text_font_style="bold"
f.xaxis.minor_tick_line_color=None
f.yaxis.minor_tick_line_color=None
f.xaxis.axis_label="Date"
f.yaxis.axis_label="Intensity"    

#create line plot, tools='pan', 
f.line(x, y)

#write the plot in the figure object
show(f)
```

```python
#making a bokeh graph with temperatures and air pressures

#importing bokeh and pandas
from bokeh.plotting import figure
from bokeh.io import output_file, show
import pandas

#prepare some data
df = pandas.read_excel("verlegenhuken.xlsx")
df["Temperature"] = df["Temperature"] / 10
df["Pressure"] = df["Pressure"] /10

#prepare the output file 
output_file("Temperatures_Pressures.html")

#creates figure object
f = figure(tools = 'pan')
f.title.text = "Temperatures and Air Pressure"
f.title.text_font = "times"
f.title.text_font_style = "bold"
f.xaxis.minor_tick_line_color = None
f.yaxis.minor_tick_line_color = None
f.xaxis.axis_label = "Tenperature (Â°C)"
f.yaxis.axis_label = "Pressure (hPa)"

#create plot, tools = 'pan'
f.circle(df["Temperature"], df["Pressure"], size = 0.5)

#write the plot in the figure object
show(f)
```

```python
from bokeh.plotting import figure, output_file, show
import pandas 

df = pandas.read_csv("adbe.csv", parse_dates = ["Date"])

p = figure(width = 500, height = 250, x_axis_type = "datetime")

p.line(df["Date"], df["Close"], color = "Orange" ,alpha = 0.5)

output_file("Timeseries.html")
show(p)
```
